```python
'''In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.
Hint:'''
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python

data = datasets.load_iris()
data
```

```python
# by seeing at the end, we came to know that its a csv file, now we read it
type(data)
```

```python
iris = data.data
iris
```

```python
data.target
```

```python
# by seeing iris, we can say that it has 4 features, now we have to transform it into 3

from sklearn.decomposition import PCA

pca = PCA(n_components = 3)

trans = pca.fit_transform(iris)

import pandas as pd

trans_df = pd.DataFrame(data = trans, columns = ['pc1','pc2','pc3'])

trans_df.head()
```

```python
# matplotlib was introduced with 2 dimensions, 3d plot is enabled by importing mplot3d toolkit, which comes with matplotlib
# instead using Axes3d(fig), we can also use this :- ax = fig.add_subplot(111, projection='3d')

# in data , we see that target is given also target names r given as target_names

from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize = (10, 10)) # creating a empty figure or plot
ax = Axes3D(fig)

x,y,z= trans_df['pc1'],trans_df['pc2'], trans_df['pc3']
# defining the axes as a 3d axes so that we can plot 3D data into it
for l in data.target_names:
    x,y,z= trans_df['pc1'],trans_df['pc2'], trans_df['pc3']
    ax.scatter(x,y,z, c= data.target, label = l)
plt.title('iris_principal_component')
ax.set_xlabel('principal_component_1')
ax.set_ylabel('principal_component_2')
ax.set_zlabel('principal_component_3')
ax.legend()
plt.show()

```
